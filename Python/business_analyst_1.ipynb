###1. Базовые подходы к анализу бизнеса.
<p> *Рынок* - это процесс непрерывного взаимодействия между продавцами товаров и услуг и потребителями.
<p>**Анализ рынка* - это исследование параметров их отношений. Продавать что? Продавать как? Продавать кому?</p>
<p> Для чего делают анализ рынка:</p>
1. Узнать потребителя - целевая аудитория.</p>
2. Разработать стратегию развития.</p>
3. Скорректировать стратегию развития.</p>
4. Для запуска проекта.</p>
5. Выйти на новый уровень дохода.</p>
<p>Наша задача: дать трезвую оценку продукта и проекта.</p>
<p>Основные моменты:</p>
1. изучить как потребители выбирают товар (услугу), на что обращают внимание.</p>
2. найти свободные ниши и выгодныен каналы продаж</p>
3. Выявить достоинства и недостатки товаров.</p>
4. Сформулировать конкурентные преимущества.</p>
5. Наметить стратегию продвижения.</p>
<p>Конечная цель анализа - соединение востребованных продуктов и увеличение прибыли.</p>
<p>Виды исследований:</p>
1. Маркетинговое. Для действующего бизнеса.</p>
2. Комплексное. Для запуска нового проекта.</p>
3. Качественное. Субьективная оценка для формирования гипотезы.</p>
4. Количественное.Основывается на количественных величинах. Статистика.</p>
